# 1. Dados <a name="data"></a>
A base de dados está disponível no kaggle: https://www.kaggle.com/olistbr/brazilian-ecommerce#olist_products_dataset.csv, em arquivos separados:

* Consumidores
* Vendedores

* Pedidos
* Artigos dos pedidos
* Avaliação de pedidos
* Pagamentos
* Geolocalização

### Bibliotecas e dependências do Python:
* Pandas – Manipulação e análise de dados 
* NumPy – Manipulação de dados  
* Seaborn - Visualização gráfica    
* Matplotlib – Visualização gráfica     

```python
#### Tratativa ################################
import pandas            as pd
import numpy             as np
import datetime          as dt
import calendar

pd.set_option('display.max_columns',500)
pd.set_option('display.max_rows',150)
pd.set_option('display.max_colwidth',150)
pd.options.display.float_format = '{:.2f}'.format

### Visualização ##############################
import matplotlib.pyplot as plt
from matplotlib          import cm
import seaborn           as sns
import folium
from folium              import plugins
from matplotlib.pyplot   import figure

plt.style.use('fivethirtyeight')
sns.set_style('darkgrid')
sns.set(font_scale = 1.2)

### Warnings #########################
import warnings
warnings.filterwarnings("ignore")

print('Bibliotecas preparadas')
```

```python
list_df = 'customers geolocation items payment orders products sellers review'.split()
```

```python
paths = {'customers'    : '../DADOS/olist_customers_dataset.csv',
         'geolocation'  : '../DADOS/olist_geolocation_dataset.csv',
         'items'        : '../DADOS/olist_order_items_dataset.csv',
         'payment'      : '../DADOS/olist_order_payments_dataset.csv',
         'orders'       : '../DADOS/olist_orders_dataset.csv',
         'products'     : '../DADOS/olist_products_dataset.csv',
         'sellers'      : '../DADOS/olist_sellers_dataset.csv',
         'review'       : '../DADOS/olist_order_reviews_dataset.csv'}

dfs = {}
for key, value in paths.items():
    dfs[key] = pd.read_csv(value)
```

```python
for df in list_df:
    print('O df', df, 'tem', dfs[df].shape[0],'linhas e ', dfs[df].shape[1], 'colunas')
```

```python
# Juntando os dfs

cus_loc = dfs['customers'].merge(dfs['geolocation'], how='inner', left_on='customer_zip_code_prefix', right_on='geolocation_zip_code_prefix').drop_duplicates('customer_id', keep='first')
cus_loc_ord = cus_loc.merge(dfs['orders'], how='inner', on='customer_id')
cus_loc_ord_item = cus_loc_ord.merge(dfs['items'], how='inner', on='order_id')
cus_loc_ord_item_prod = cus_loc_ord_item.merge(dfs['products'], how='inner', on='product_id')
cus_loc_ord_item_prod_revi= cus_loc_ord_item_prod.merge(dfs['review'], how='left', on='order_id')
cus_loc_ord_item_prod_revi_pay = cus_loc_ord_item_prod_revi.merge(dfs['payment'], how='left', on='order_id')

df = cus_loc_ord_item_prod_revi_pay.copy()

# Convertendo para datetime

colunas_datas = ['order_purchase_timestamp',
           'order_approved_at',
           'order_delivered_carrier_date',
           'order_delivered_customer_date',
           'order_estimated_delivery_date',
           'shipping_limit_date',
           'review_creation_date',
           'review_answer_timestamp']

for data in colunas_datas:
     df[data] = pd.to_datetime(df[data])

# Criando coluna de tempo de entrega e Hora da compra

df['tempo_entrega'] = (df['order_delivered_customer_date'].dt.date - df['order_purchase_timestamp'].dt.date).dt.days
df['dia_compra'] = df['order_purchase_timestamp'].dt.day
df['mes_compra'] = df['order_purchase_timestamp'].dt.month
df['ano_compra'] = df['order_purchase_timestamp'].dt.year
df['nome_dia_compra'] = df['order_purchase_timestamp'].dt.day_name()
df['hora_compra'] = df['order_purchase_timestamp'].dt.hour

# Traduzindo os dias da semana

dic_dias = {'Tuesday'    : 'Terça',
           'Thursday'   : 'Quinta',
           'Friday'     : 'Sexta',
           'Sunday'     : 'Domingo',
           'Monday'     : 'Segunda',
           'Saturday'   : 'Sábado',
           'Wednesday'  : 'Quarta'}

df['nome_dia_compra'] = df['nome_dia_compra'].map(dic_dias)

del cus_loc, cus_loc_ord, cus_loc_ord_item, cus_loc_ord_item_prod, cus_loc_ord_item_prod_revi, cus_loc_ord_item_prod_revi_pay
```

# 3. Análises Descritivas <a name="analises"></a>

---
### 3.1 Vendas por mês

**Gráfico I**

No ano de 2016 temos pouco registros, nenhuma conclusão pode ser feita.

No ano de 2017 podemos ver um aumento nas vendas no mês 11 (Novembro), podendo considerar que ocorreu devido ao evento **Black Friday**.

Em 2018 as vendas em todos os mês aumentaram pelo menos 30%, em relação ao ano de 2017.

```python
for ano in sorted(df['ano_compra'].unique()):
    
    plt.figure(figsize=(15,6))
    
    dic_color={2016:'lightcoral', 2017:'turquoise', 2018:'mediumpurple'}
    
    df.loc[df['ano_compra']==ano,'mes_compra'].value_counts().sort_index().plot.bar(color=dic_color[ano])
    
    plt.title('Gráfico I - Total de vendas por mês no ano de {}'.format(ano), fontsize=16)
    plt.xlabel('Ano')
    plt.ylabel('Total de vendas')
```

---
### 3.2 Quantidade de Compras por Horário
**Gráfico II**. É possível constatar que o período de maior movimentação de compras é compreendido entre 10:00 e 22:00 entre segunda e sexta. Aos finais de semana, o movimento é menor.

```python
dia_hora=df.groupby(['nome_dia_compra','hora_compra']).agg({'order_id':'count'}).rename(columns={'order_id':'freq'}).reset_index()
dia_hora = dia_hora.pivot('nome_dia_compra','hora_compra','freq')
dia_hora = dia_hora.fillna(0)
dia_hora = dia_hora.astype(int)
dia_hora = dia_hora.loc[['Domingo','Segunda','Terça','Quarta','Quinta','Sexta','Sábado'],:]

plt.figure(figsize=(15,8))
sns.heatmap(dia_hora,annot=True,fmt='d',cmap="OrRd",linewidths=.1, cbar=False)
plt.xlabel("Hora")
plt.ylabel("Dia")
plt.title("Gráfico II - Mapa de calor de compras por dia da semana e hora",size=16)
plt.yticks(va='center')

plt.show()
```

---
### 3.3 Quantidade de Clientes por Estado

**Gráfico III**.
É possível constatar que 69.93% dos clientes estão localizados na região Sudeste do Brasil.

```python
plt.figure(figsize=(15,6))   

color=cm.get_cmap('Blues', 28)(np.linspace(0, 1, 28))[::-1]

df['customer_state'].value_counts().plot.bar(color=color)

plt.title('Gráfico III - Dispersão de clientes por estado', fontsize=16)
plt.xlabel('Estado')
plt.ylabel('Quantidade de Clientes')

plt.show()

for estado in df['customer_state'].value_counts().head().index:
    print(str(round((sum(df['customer_state']==estado)/df.shape[0])*100,2))+'% dos clientes no estado de {}'.format(estado))
```

---
### 3.4 Produtos mais vendidos
**Gráfico IV**. É possível constatar que produtos de **cama_mesa_banho**, **beleza_saude** são os mais vendidos.

```python
plt.figure(figsize=(15,6))

df['product_category_name'].value_counts().head(10).plot.bar(color = cm.get_cmap('Blues', 15)(np.linspace(0, 1, 15))[::-1])

plt.title('Gráfico IV - Produtos', fontsize=16)
plt.xlabel('Produtos')
plt.ylabel('Quantidade de vendas')
plt.xticks(rotation=45, ha='right')

plt.show()
```

___
### 3.5 Quantidade de notas por score


**Gráfico V**. Podemos observar que 73.75% dos clientes que utilizam o e-commerce, realizam o pagamento via cartão de crédito.

```python
plt.figure(figsize=(15,6))

df['payment_type'].value_counts().plot.bar(color=color)

plt.title('Gráfico V - Dispersão de compras por ordem de pagamento', fontsize=16)
plt.xlabel('Forma de Pagamento')
plt.ylabel('Quantidade')

plt.show()

for tipo in df['payment_type'].value_counts().index:
    print(str(round((sum(df['payment_type']==tipo)/df.shape[0])*100,2))+'% dos clientes que fizeram a forma de pagamento como: {}'.format(tipo))
```

___
### 3.6 Quantidade de notas por score

**Gráfico VI**. É possivel observar que, 74.98% das pessoas que compram no e-commerce no Brasil, classificam suas compras com notas 4 e 5.

```python
for score in df['review_score'].value_counts().index:
    print(str(round((sum(df['review_score']==score)/df.shape[0])*100,2))+'% dos clientes classificaram suas compras com nota {}'.format(score))

color=cm.get_cmap('Blues', 6)(np.linspace(0, 1, 6))[::-1]

df['review_score'].value_counts().plot.bar(figsize=(15,6), color=color)

plt.title('Gráfico VI - Dispersão de review por score', fontsize=16)
plt.xlabel('Score')
plt.ylabel('Votos')

plt.xticks(rotation=0)

plt.show()
```

# 4. Conclusão <a name="conclusao"></a>

* Embora não tenhamos muitos registros afim de observar uma sazonalidade no consumo, podemos notar que a partir do ano de 2018 as vendas aumentaram 30% em relação ao ano anterior.


* Maior parte das compras são realizadas de segunda a sexta, entre as 10:00 e às 22:00.


* 70% das vendas no e-commerce no Brasil, são feitas pela região Sudeste (São Paulo, Rio de Janeiro e Minas Gerais).


* 73.75% dos clientes, fazem o pagamento com cartão de crédito.


* 75% das vendas foram classificadas com nota 5, podendo concluir que os clientes estão satisfeitos com as compras pelo e-commerce, pela facilidade no pagamento, entregas no prazo e sem muitas complicações. 


* Grande parte das insatisfações dos clientes, são devido a demora nas entregas, entregas parciais ou não entrega dos produtos.
